```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Write a Python program to visualize the data from the Dataframe

a) Line Chart , b) Bar Graph c) Histogram
show legends and different colours.

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

# Question 1: Line Chart, Bar Graph, and Histogram of the Dataframe
# a) Line Chart
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Open'], label='Open', color='blue')
plt.plot(df['Date'], df['High'], label='High', color='green')
plt.plot(df['Date'], df['Low'], label='Low', color='red')
plt.plot(df['Date'], df['Close'], label='Close', color='purple')
plt.title('Line Chart of Stock Data')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# b) Bar Graph
df.set_index('Date')[['Open', 'High', 'Low', 'Close']].plot(kind='bar', figsize=(10, 6))
plt.title('Bar Graph of Stock Data')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# c) Histogram
df[['Open', 'High', 'Low', 'Close']].plot(kind='hist', alpha=0.7, bins=10, figsize=(10, 6), color=['blue', 'green', 'red', 'purple'])
plt.title('Histogram of Stock Data')
plt.xlabel('Price')
plt.legend()
plt.show()
```

2. Read all columns of the dataframe and show it using a multiline plot

```python
# Multiline Plot
plt.figure(figsize=(12, 6))

# Plotting all columns
plt.plot(df.index, df['Open'], label='Open', color='blue')
plt.plot(df.index, df['High'], label='High', color='green')
plt.plot(df.index, df['Low'], label='Low', color='red')
plt.plot(df.index, df['Close'], label='Close', color='purple')

# Adding titles and labels
plt.title('Stock Prices Over Time - Multiline Plot')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.show()
```

```python
df.plot(x='Date', y=['Open', 'High', 'Low', 'Close'], figsize=(10, 5))
plt.title('Multiline Plot of All Columns')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Question 6: Close Column Visualization
# a) Histogram
plt.figure(figsize=(10, 5))
plt.hist(df['Close'], bins=10, color='purple', alpha=0.7)
plt.title('Histogram of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Reset index if necessary
df = df.reset_index()  # Uncomment if 'Date' is already the index

plt.figure(figsize=(10, 5))
plt.bar(df['Date'].dt.strftime('%m-%d'), df['Close'], color='purple')
plt.title('Bar Chart of Close Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```

```python
# c) Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(df['Close'], labels=df['Date'].dt.strftime('%m-%d'), autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Pie Chart of Close Prices')
plt.show()
```
