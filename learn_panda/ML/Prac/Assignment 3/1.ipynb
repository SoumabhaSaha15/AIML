```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create DataFrame
df = pd.DataFrame(data={
    'a': np.random.randint(0, 100, 30),
    'b': np.random.randint(0, 100, 30),
    'c': np.random.randint(0, 100, 30)
})
```

```python
# Create a figure with size 15x8 and two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))

# Top subplot
ax1.plot(df.index, df['a'], color='green', label='Green')
ax1.plot(df.index, df['b'], color='orange', label='Orange')
ax1.legend(loc='upper center')
ax1.set_title('Top Plot: Green and Orange Lines')
ax1.set_xlabel('Index')
ax1.set_ylabel('Values')

# Bottom subplot
ax2.scatter(df.index, df['a'], color='blue', marker='o', label='a')
ax2.scatter(df.index, df['b'], color='red', marker='o', label='b')
ax2.scatter(df.index, df['c'], color='purple', marker='o', label='c')
ax2.set_title('Bottom Plot: Data Points Only')
ax2.set_xlabel('Index')
ax2.set_ylabel('Values')
ax2.legend(loc='upper right')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Bar graph
df.plot(kind='bar', figsize=(12, 6), colormap='viridis')
plt.title('Bar Graph of Data')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
