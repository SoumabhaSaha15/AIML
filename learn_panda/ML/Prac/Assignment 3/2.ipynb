Exercise 2 : According to the dataframe in Exercise 1:  

a)Create a figure with three subplots, aligned vertically. 

b)Create a bar graph for a and b on the first two, 

c) Create a stacked bar graph by putting b on top of a Design a piechart for a and b

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create DataFrame
df = pd.DataFrame(data={
    'a': np.random.randint(0, 100, 30),
    'b': np.random.randint(0, 100, 30),
    'c': np.random.randint(0, 100, 30)
})
```

```python
# a) Create a figure with 3 subplots, aligned vertically
fig, axs = plt.subplots(3, 1, figsize=(10, 7))

# Bar graph for 'a' on the first subplot
axs[0].bar(df.index, df['a'], color='blue', label='a')
axs[0].set_title('Bar Graph of a')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Values')

# Bar graph for 'b' on the second subplot
axs[1].bar(df.index, df['b'], color='green', label='b')
axs[1].set_title('Bar Graph of b')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Values')

# Stacked bar graph with 'b' on top of 'a' on the third subplot
axs[2].bar(df.index, df['a'], color='blue', label='a')
axs[2].bar(df.index, df['b'], bottom=df['a'], color='green', label='b')
axs[2].set_title('Stacked Bar Graph of a and b')
axs[2].set_xlabel('Index')
axs[2].set_ylabel('Values')
axs[2].legend()

# Adjust layout to prevent overlap
plt.tight_layout()
```

```python
# Create a new figure for pie charts
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# Pie chart for 'a'
axs[0].pie(df['a'], labels=df.index, autopct='%1.1f%%', colors=plt.cm.Blues(np.linspace(0, 1, len(df['a']))))
axs[0].set_title('Pie Chart of a')

# Pie chart for 'b'
axs[1].pie(df['b'], labels=df.index, autopct='%1.1f%%', colors=plt.cm.Greens(np.linspace(0, 1, len(df['b']))))
axs[1].set_title('Pie Chart of b')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```
