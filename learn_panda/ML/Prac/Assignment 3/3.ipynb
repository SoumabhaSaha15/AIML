Language = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'] 
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7] 
Write a Python programming to display a  
a) Bar chart of the popularity of programming Languages. Use different color for each bar. 

b) Create a group barchart by taking two consecutive elements from the language list to form a group. 

c) Create a stacked barchart for the groups. 
( Visualization styles : Provide suitable label in the x axis, y axis and a title , legends , and border line) 
 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
# Create a new figure
fig, axs = plt.subplots(3, 1, figsize=(12, 18))

# a) Bar Chart of the Popularity of Programming Languages
colors = plt.cm.tab10(np.linspace(0, 1, len(languages)))  # Different colors for each bar
bars = axs[0].bar(languages, popularity, color=colors, edgecolor='black')

# Add labels and title
axs[0].set_xlabel('Programming Languages')
axs[0].set_ylabel('Popularity (%)')
axs[0].set_title('Popularity of Programming Languages')
axs[0].legend(bars, languages, title='Languages', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add border line
for bar in bars:
    bar.set_edgecolor('black')
    
    
# b) Grouped Bar Chart
bar_width = 0.35  # Width of each bar
index = np.arange(len(languages) // 2)  # Positions for the groups

# Data for groups
for i in range(0, len(languages), 2):
    axs[1].bar(index[i // 2] - bar_width / 2, popularity[i], bar_width, label=languages[i], edgecolor='black')
    axs[1].bar(index[i // 2] + bar_width / 2, popularity[i + 1], bar_width, label=languages[i + 1], edgecolor='black')

# Add labels and title
axs[1].set_xlabel('Groups of Programming Languages')
axs[1].set_ylabel('Popularity (%)')
axs[1].set_title('Grouped Bar Chart of Programming Languages')
axs[1].set_xticks(index)
axs[1].set_xticklabels([f'{languages[i]} & {languages[i+1]}' for i in range(0, len(languages), 2)])
axs[1].legend(title='Languages')

# Add border line
for patch in axs[1].patches:
    patch.set_edgecolor('black')
    
    
# c) Stacked Bar Chart
index = np.arange(len(languages) // 2)
bars_a = np.zeros(len(index))
bars_b = np.zeros(len(index))

# Stacking bars
for i in range(0, len(languages), 2):
    bars_a[i // 2] = popularity[i]
    bars_b[i // 2] = popularity[i + 1]

# Plot stacked bars
axs[2].bar(index, bars_a, bar_width, label='Language 1', color='blue', edgecolor='black')
axs[2].bar(index, bars_b, bar_width, bottom=bars_a, label='Language 2', color='green', edgecolor='black')

# Add labels and title
axs[2].set_xlabel('Groups of Programming Languages')
axs[2].set_ylabel('Popularity (%)')
axs[2].set_title('Stacked Bar Chart of Programming Languages')
axs[2].set_xticks(index)
axs[2].set_xticklabels([f'{languages[i]} & {languages[i+1]}' for i in range(0, len(languages), 2)])
axs[2].legend(title='Languages')

# Add border line
for patch in axs[2].patches:
    patch.set_edgecolor('black')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
